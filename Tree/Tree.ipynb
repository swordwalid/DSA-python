### Tree Implementation

```python
class Tree:

  def __init__(self,name=None):
    self.name=name
    self.children=[]
    self.parent=None

  def add_child(self,child):
    child.parent=self
    self.children.append(child)

  def get_level(self):
    level=0
    p=self.parent
    while p:
      level+=1
      p=p.parent
    return level


  def print_whole_tree(self):
    spaces = ' ' * self.get_level() * 3
    prefix = spaces + "|__" if self.parent else ""
    print(prefix + self.name)
    if self.children:
      for child in self.children:
        child.print_whole_tree()



```

```python
root=Tree("Electronics")

Laptops=Tree("Laptops")
Laptops.add_child(Tree("Macbook"))
Laptops.add_child(Tree("Microsoft surface"))
Laptops.add_child(Tree("Thinkpad"))

cellphones=Tree("cellphones")
cellphones.add_child(Tree("iphone"))
cellphones.add_child(Tree("Tecno"))
cellphones.add_child(Tree("Vivo"))

Tv=Tree("Televisions")
Tv.add_child(Tree("Samsung"))
Tv.add_child(Tree("LG"))
Tv.add_child(Tree("Hisense"))

root.add_child(Laptops)
root.add_child(cellphones)
root.add_child(Tv)

root.print_whole_tree()
```

Extent tree class that is already buit in so that it takes name and designation in data part of TreeNode class. Now extend print_tree function such that it can print either name tree, designation tree or name and designation tree. As shown below,

all_trees.png

```python

class ManagementHierarchy(Tree):
  def __init__(self,name,designation=None):
    super().__init__(name)
    self.designation=designation

  def get_spaces(self):
    spaces=" "*self.get_level()
    prefix=spaces + "|__" if self.parent else ""
    return prefix

  def print_whole_tree(self,status):
        # Print designations only
        if status=="designation":
          self._print_designation()
          print("---------------------")

        # Print names only
        if status=="name":
          self._print_name()
          print("---------------------")

        # Print designations and names
        if status=="both":
          self._print_designation_and_name()
          print("---------------------")

  def _print_designation(self):
    print(self.get_spaces() + self.designation)
    for child in self.children:
      child._print_designation()

  def _print_name(self):
    print(self.get_spaces() + self.name)
    for child in self.children:
      child._print_name()

  def _print_designation_and_name(self):
    print(self.get_spaces() + self.designation + " (" + self.name + ")")
    for child in self.children:
      child._print_designation_and_name()
```

```python
root=ManagementHierarchy("Nilupul","CEO")

cto=ManagementHierarchy("Chinmay","CTO")
ihc=ManagementHierarchy("Vishwa","Infrastucture Head")
ihc.add_child(ManagementHierarchy("Dhaval","cloud Manager"))
ihc.add_child(ManagementHierarchy("Abhijit","App Manager"))
cto.add_child(ihc)
cto.add_child(ManagementHierarchy("Aamir","Application Head"))

hrh=ManagementHierarchy("Gels","HR Head")
hrh.add_child(ManagementHierarchy("Peter","Recruitment Manager"))
hrh.add_child(ManagementHierarchy("Waqas","Policy Manager"))

root.add_child(cto)
root.add_child(hrh)
```

Now modify print_tree method to take tree level as input. And that should print tree only upto that level as shown below,

```python
class Tree2(Tree):

  def __init__(self,name):
    super().__init__(name)

  def print_whole_tree(self, level):
    spaces = ' ' * self.get_level() * 3
    prefix = spaces + "|__" if self.parent else ""
    print(prefix + self.name)
    for child in self.children:
      if child.get_level() <= level:
        child.print_whole_tree(level)


```

```python
root_node=Tree2("Global")

india=Tree2("India")
gujrat=Tree2("Gujarat")
gujrat.add_child(Tree2("Ahmedabad"))
gujrat.add_child(Tree2("Baroda"))

karnataka=Tree2("Karnataka")
karnataka.add_child(Tree2("Bangluru"))
karnataka.add_child(Tree2("Mysore"))

india.add_child(gujrat)
india.add_child(karnataka)

usa=Tree2("USA")

new_jersey=Tree2("New Jersey")
new_jersey.add_child(Tree2("Princeton"))
new_jersey.add_child(Tree2("Trenton"))

california=Tree2("California")
california.add_child(Tree2("San Francisco"))
california.add_child(Tree2("Mountain View"))
california.add_child(Tree2("Palo Alto"))

usa.add_child(new_jersey)
usa.add_child(california)

root_node.add_child(india)
root_node.add_child(usa)
```

### Binary Tree

```python
class BinaryTree:

    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


    def add_child(self, child):
        if child.data == self.data:
            return

        if child.data < self.data:
            if self.left:
                self.left.add_child(child)
            else:
                self.left = child
        else:
            if self.right:
                self.right.add_child(child)
            else:
                self.right = child

    def in_order_traversal_for_opereation(self,entry=None):
        if entry is None:
          entry=[]

        if not self:
            return
        if self.left:
            self.left.in_order_traversal_for_opereation(entry)
        entry.append(self.data)
        if self.right:
            self.right.in_order_traversal_for_opereation(entry)
        return entry

    def in_order_traversal(self):
        if not self:
            return
        if self.left:
            self.left.in_order_traversal()
        print(self.data)
        if self.right:
            self.right.in_order_traversal()


    def pre_order_traversal(self):
        if not self:
            return
        print(self.data)
        if self.left:
            self.left.pre_order_traversal()
        if self.right:
            self.right.pre_order_traversal()

    def post_order_traversal(self):
        if not self:
            return
        if self.left:
            self.left.post_order_traversal()
        if self.right:
            self.right.post_order_traversal()
        print(self.data)

    def find_min(self):
      result=self.in_order_traversal_for_opereation()
      return min(result)

    def find_max(self):
      result=self.in_order_traversal_for_opereation()
      return max(result)

    def calculate_sum(self):
      return sum(self.in_order_traversal_for_opereation())

    def delete(self,data):
      if not self: return None

      if self.data==data:
        if not self.left and not self.right: return None
        if not self.left and self.right: return self.right
        if self.left and not self.right return self.left
        # if both:
        pointer=self.right
        while pointer.left:pointer=pointer.left     # to get the successor that is grater than the current node
        self.data=pointer.data
        self.right=self.delete(self.right,data)
      if data<self.data:
        self.left=self.delete(self.left,data)
      if data> self.data:
        self.right=self.delete(self.right,data)




```

```python
root=BinaryTree(15)

order_1_left=BinaryTree(12)
order_1_left.add_child(BinaryTree(7))
order_1_left.add_child(BinaryTree(14))

root.add_child(order_1_left)
order_2_left=BinaryTree(20)
order_1_right=BinaryTree(27)
order_1_right.add_child(order_2_left)
order_2_left.add_child(BinaryTree(23))
order_1_right.add_child(BinaryTree(88))
root.add_child(order_1_right)
```
